Second, is that you can make a set by instantiating one using `Set()`.

````python
d = [1,2,3,4,5]
````

````python
e = Set(d)
````

Third, you can instantiate a set using a list as a parameter in the set's constructor call, as in cell 13 & a4 above.

