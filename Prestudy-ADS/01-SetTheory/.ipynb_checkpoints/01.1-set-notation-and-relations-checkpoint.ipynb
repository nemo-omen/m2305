# Sets
A set is generally understood to be a collection of unique elements.

Sets are named like variables. When a set is given a letter name, like `A`, that name is uppercase by convention.

If elements in a set are referred to using a letter variable, those variable names are in lowercase, like `x`.

## Common sets

$\Bbb P$: the positive integers $\{1,2,3,4,5,...\}$

$\Bbb N$: the natural numbers $\{0,1,2,3,4,...\}$

$\Bbb Z$: the integers $\{...,-3,-2,-1,0,1,2,3,...\}$

$\Bbb Q$: the rational numbers $\{1/2,1/3,1/4,...\}$

$\Bbb R$: the real numbers

$\Bbb C$: the complex numbers

$\emptyset$: the empty set

## Set Builder Notation

$\Bbb Q=\{a/b \mid a,b \in \Bbb Z, b \neq 0 \}$

(translation: the set $Q$ is equal to all rational numbers, $a/b$ such that $a$ and $b$ are integers and $b$ does not equal $0$)

## Sets and their members

Let's say we have a set whose elements are $x, y, \text{ and } z$, we can write that with the following notation:

$A=\{x,y,z\}$

To show that one element is a member of $A$, we write:

$x\in A $

To show that a given value or element is _not_ a member of the set, we write:

$w \notin A $

### Finite Sets, Infinite Sets

A set is a finite set if it has a finite number of elements. Any set that is not finite is an infinite set.

### Cardinality
Let $A$ be a _finite set_. 

The number of different elements in $A$ is called its __cardinality__.

The cardinality of finite set $A$ is denoted $\mid A \mid$.

__Ex:__ $A = \{1,2,3\}$, $\mid A \mid = 3$

_(There are also sets with infinite cardinality, but that will come later)_

## Subsets

Let $A$ and $B$ be sets. 

$A$ is a __subset__ of $B$ if and only if every element of $A$ is also an element of $B$. 

This is denoted as $A \subseteq B$.

__Ex:__ $A=\{1,3,5\}$, $B=\{1,2,3,4,5,6\}$

$A\subseteq B$

$\Bbb N \subseteq \Bbb Z \subseteq \Bbb Q \subseteq \Bbb R \subseteq \Bbb C$

If $S = \{1,5,3\}$ and $T=\{5,3,1\}$ then $S \subseteq T$ and $T \subseteq S$

#### Proper and Improper subsets

Given a set $A$, $A \subseteq A$ and $\emptyset \subseteq A$. If $A$ is nonempty, then $A$ is called an __improper subset__ of $A$.

All other subsets of $A$ are a __proper subset__ of $A$, (denoted using $\subset$).

$\emptyset$ is an improper subset of itself. ($\emptyset \subseteq \emptyset$)

__Ex:__ $A = \{1,2,3,4,5\}$, $B=\{1,2\}$, $C=\{1,2,3,4,5\}$

$A \subseteq C$, $C \subseteq A$, $B \subset A$, $B \subset C$

## Set Equality

Let $A$ and $B$ be sets. $A$ is equal to $B$ ($A=B$) if and only if every element in $A$ is in $B$ __and__ every element in $B$ is also in $A$.

$A=B$ if $A\subseteq B$ and $B\subseteq$

$A=\{1,5,3,5\}$, $B=\{1,5,3\}$

$B=A$


(This text does not deal with __multisets__, sets in which the number of elements matters.)

$S=\{3,5,8\}$, $T=\{5,8,3\}$

$S\subseteq T$, $T \subseteq S$, $S=T$
