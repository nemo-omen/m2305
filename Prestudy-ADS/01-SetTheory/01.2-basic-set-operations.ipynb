# Basic Set Operations

## Definitions

### Intersection

Let $A$ and $B$ be sets. The __intersection__ of $A$ and $B$ (denoted $ A \cap B $) is the set of all elements that are in both $A$ and $B$.

$ A \cap B = \{x \mid x \in A, x \in B  \} $  

### Disjoint Sets
Two sets are __disjoint__ if they have no elements in common. $A$ and $B$ are disjoint if $A \cap B = \emptyset$.

__EX:__

Let $A = \{ 1,3,8 \} $ and $B= \{ -9,22,3 \}$. Then $ A \cap B = \{ 3 \} $ 

$\Bbb Z \cap \Bbb Q = \Bbb Z$

If $A = \{3,5,9\}$ and $ B = \{-5,8\}$, then $A\cap B = \emptyset$

### Union

Let $A$ and $B$ be sets. The __union__ of $A$ and $B$, (denoted $A \cup B$), is the set of all elements that are in $A$, in $B$, or in both.

$ A \cup B = \{x \mid x \in A \text{ or } x \in B \}  $

__EX:__

if $A= \{ 2,5,8 \}$ and $B= \{ 7,5,22 \}$, then $A\cup B= \{ 2,5,7,8,22\}$

$ Z \cup Q = \{ Q \} $

$ A \cup \emptyset = A \text{ for any set } A$

### Universe

The __universe__, or __universal set__, is the set of all elements under discussion for possible membership in a set. The letter $U$ is normally reserved for representing the universe set.
