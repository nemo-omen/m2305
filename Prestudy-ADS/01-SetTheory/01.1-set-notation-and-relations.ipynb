# Sets
A set is generally understood to be a collection of unique elements.

Sets are named like variables. When a set is given a letter name, like `A`, that name is uppercase by convention.

If elements in a set are referred to using a letter variable, those variable names are in lowercase, like `x`.

## Common sets

$\Bbb P$: the positive integers $\{1,2,3,4,5,...\}$

$\Bbb N$: the natural numbers $\{0,1,2,3,4,...\}$

$\Bbb Z$: the integers $\{...,-3,-2,-1,0,1,2,3,...\}$

$\Bbb Q$: the rational numbers $\{1/2,1/3,1/4,...\}$

$\Bbb R$: the real numbers

$\Bbb C$: the complex numbers

## Set Builder Notation

$\Bbb Q=\{a/b \mid a,b \in \Bbb Z, b \neq 0 \}$

(translation: the set $Q$ is equal to all rational numbers, $a/b$ such that $a$ and $b$ are integers and $b$ does not equal $0$)

## Sets and their members

Let's say we have a set whose elements are $x, y, \text{ and } z$, we can write that with the following notation:

$A=\{x,y,z\}$

To show that one element is a member of $A$, we write:

$x\in A $

To show that a given value or element is _not_ a member of the set, we write:

$w \notin A $
